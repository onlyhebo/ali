```python
import numpy as np
import pandas as pd
from collections import Counter
#num = 50000
train_X = open('.././input/train_X.txt','r',encoding = 'utf8')
train_Y = open('.././input/train_Y.txt','r',encoding = 'utf8')
X_read = train_X.readlines()
Y_read = train_Y.readlines()
for i in range(len(X_read)):
    X_read[i] = X_read[i].split('\n')[0]
    Y_read[i] = Y_read[i].split('\n')[0]
#len(X_read) = 116385
```

## 挑选出次数在5到15之间的词

```python
X = []
Y = []
for i in range(len(X_read)):
    if len(X_read[i].split()) >= 5:
        X.append(X_read[i])
        Y.append(Y_read[i])
#len(X) = 109284
```

## label再分词结果里面的

```python
X_refresh = []
Y_refresh = []
for i in range(len(X)):
    if Y[i].split('\n')[0] in X[i].split():
        if len(Y[i].split('\n')[0]) != 1:
            X_refresh.append(X[i])
            Y_refresh.append(Y[i])   
#Y_refresh = 65560
```

## 创建规则字典。清洗打标不一致的现象。

```python
label_dict = {} # 创建规则字典
for i in range(len(X_refresh)):
    replace = False 
    label_cadi = [] #找出备选标签
    for word in X_refresh[i].split():
        if word in Y_refresh:
            label_cadi.append(word)
    #将备选标添加到label_cadi里面
    if Y_refresh[i] not in label_dict.keys():
        label_dict[Y_refresh[i]] = []
    for word_1 in label_cadi:
        for word_2 in label_cadi:
            if word_1 != word_2:
                if word_1 in label_dict.keys():
                    if word_2 in label_dict[word_1]:
                        Y_refresh[i] = word_1
                        replace = True
                        break
                #如果规则字典里面有，将此时Y值更新为规则词典
        if replace == True:
            break
        #如果已经代替，跳出循环
    if replace == True:
        continue
    #如果已经代替。继续下一条商品描述
    else:
        #label_cadi.remove(Y_refresh[i])
        #label_dict[Y_refresh[i]].extend(label_cadi)
        for word in label_cadi:
            if word!= Y_refresh[i]:
                label_dict[Y_refresh[i]].append(word)
     #如果没有代替的话，创建该规则字典   
#可以考虑加入if Y_refresh.count(Y_refresh[i]) < 10  直接break
```

## 统计label数目大于10的数据

```python
X_new = []
Y_new = []
for i in range(len(Y_refresh)):
    if Y_refresh.count(Y_refresh[i]) >= 50 and Y_refresh.count(Y_refresh[i]) <= 500:
        X_new.append(X_refresh[i])
        Y_new.append(Y_refresh[i])
for i in range(len(X_new)):
    if Y_new[i] == '短袖t恤':
        Y_new[i] = '短袖T恤'
    if Y_new[i] == 't恤':
        Y_new[i] = 'T恤'
#len(X_new)= 44528  label数目大于等于10的
#len(X_new) = 47136 lbael 数目大于等于8的
```

```python
with open('X_new.txt','w',encoding = 'utf8') as X_op:
    for i in range(len(X_new)):
        X_op.write(X_new[i] + '\n')
with open('Y_new.txt','w',encoding = 'utf8') as Y_op:
    for i in range(len(Y_new)):
        Y_op.write(Y_new[i] + '\n')
```

## word2vec

```python
with open('X_new.txt','w',encoding = 'utf8') as op:
    for i in range(len(X_new)):
        op.write(X_new[i] + '\n')
from gensim.models import word2vec
sentences = word2vec.Text8Corpus('X_new.txt')
model = word2vec.Word2Vec(sentences,size = 100,window = 3 ,min_count = 3)
model.wv.save_word2vec_format('wiki.zh.text.vector',binary =False)
```

```python
output = open('word2cev_50dim.txt','w',encoding = 'utf8')
misnum = 0
dim = 100
for line in range(len(X_new)):
    sum = np.zeros((1,dim))
    words = X_new[line].split()
    num = 1
    for word in words:
        try:
            temp = model[word]
        except:
            misnum +=1
        else:
            temp = temp.reshape((1,dim))
            sum += temp
            num = num+1
    if num > 1:
        num -= 1
    sum /= num
    for j in range(dim):
        output.write('%.4f'%(sum[0][j]) + ' ')
    output.write('\n')
output.flush() #刷新缓冲区
output.close()
```
