```python
import jieba
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score,accuracy_score
from sklearn.cross_validation import train_test_split
import numpy as np
import pandas as pd 
jieba.add_word('caonidaye')
X = [line.strip('\n') for line in open('.././input/train_X.txt','r',encoding = 'utf8').readlines()]
Y = [line.strip('\n') for line in open('.././input/train_Y.txt','r',encoding = 'utf8').readlines()]
#筛选词不在商品里面的 筛除 116358 - 104252 = 12096 条数据
#X_del用来存放这些垃圾
```

```python
def fenciClean(X_1,Y_1):
    X_2 = []
    Y_2 = []
    X_del = []
    Y_del = []
    for i in range(len(Y_1)):
        description = X_1[i].replace(' ','')
        if Y_1[i] in description and Y_1[i] != '':
            drop_Y = description.replace(Y_1[i],'caonidaye') # 将描述里面的关键词抽取出来
            drop_cut = jieba.lcut(drop_Y) #对剩余的商品描述分词
            s = ''
            for j in range(len(drop_cut)):
                if drop_cut[j] == 'caonidaye':
                    drop_cut[j] = Y_1[i]
                s = s + drop_cut[j] + ' '
            X_2.append(s)
            Y_2.append(Y_1[i])
        else:
            X_del.append(X_1[i])
            Y_del.append(Y_1[i])
    X_3 = []
    Y_3 = []
    for i in range(len(X_2)):
        if len(X_2[i].split())>=3:
            X_3.append(X_2[i])
            Y_3.append(Y_2[i])
    return X_3,Y_3
```

```python
def choose(x,y,l,h):
    dic = Counter(y)
    x_1 = []
    y_1 = []
    for i in range(len(x)):
        if dic[y[i]] >= l and dic[y[i]] <= h:
            x_1.append(x[i])
            y_1.append(y[i])
    print(len(y_1),len(Counter(y_1).keys()))
    return x_1,y_1
def LR_countVector(X,Y):
    vectorizer = CountVectorizer()
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
    X_train_1 = vectorizer.fit_transform(X_train + X_test)
    classifier = LogisticRegression(random_state = 0)
    classifier.fit(X_train_1[:len(X_train)], Y_train)# coding:utf-8
    y_pred = classifier.predict(X_train_1[len(X_train):])
    p = np.mean(y_pred == Y_test)
    print(p)
    return X_test,y_pred,Y_test
def LR_Tfidf(X,Y):
    tfidfVecorizer  = TfidfVectorizer()
    tfidf = tfidfVecorizer.fit_transform(X)
    X_train,X_test,Y_train,Y_test = train_test_split(tfidf,Y,test_size = 0.2)
    clf = LogisticRegression()
    clf.fit(X_train,Y_train)
    pre = clf.predict(X_test)
    print(np.mean(pre == Y_test))
    return pre,Y_test
def Accuracy(predict,true):
    a = accuracy_score(true, predict)
    p = precision_score(true, predict, average='macro')  
    r = recall_score(true, predict, average='macro')  
    f1score = f1_score(true, predict, average='macro')
    print(a)
    print(p)  
    print(r)  
    print(f1score)  
def ErrorName(X_test,Y_pre,Y_true):
    for i in range(len(Y_pre)):
        if Y_pre[i] != Y_true[i]:
            print(X_test[i])
            print(Y_pre[i],Y_true[i])
```

```python
def Rule(X_fenci,Y_fenci):
    label_dict = {} # 创建规则字典
    del_index = []
    for i in range(len(X_fenci)):
        if i % 1000 == 0:
            print(i)
        replace = False
        delete = False
        label_cadi = [] #找出备选标签
        for word in X_fenci[i].split():
            if word in Y_fenci:
                label_cadi.append(word)
        label_cadi = list(set(label_cadi))
        label_cadi.remove(Y_fenci[i])
        if len(label_cadi) == 0:
            continue
        if Y_fenci[i] not in sum(list(label_dict.values()),[]):
            if Y_fenci[i] not in label_dict.keys():
                label_dict[Y_fenci[i]] = label_cadi
            else:
                label_dict[Y_fenci[i]] = list(set(label_dict[Y_fenci[i]] + label_cadi))
        else:
            ti = True
            n = 0  
            while ti:
                times = len(label_cadi)
                for j in range(len(label_cadi)):
                    if label_cadi[j] in label_dict.keys():
                        if Y_fenci[i] in label_dict[label_cadi[j]]:
                            temp = Y_fenci[i]
                            Y_fenci[i] = label_cadi[j]
                            #label_cadi.remove(label_cadi[j])
                            label_cadi[j] = temp
                            replace = True
                            break
                n += 1
                if n == times + 1:
                    del_index.append(i)
                    delete = True
                    ti = False
                if j == len(label_cadi) - 1:
                    ti = False
        if delete == True:
            continue
        else:
            if Y_fenci[i] not in label_dict.keys():
                label_dict[Y_fenci[i]] = label_cadi
            else:
                label_dict[Y_fenci[i]] = list(set(label_dict[Y_fenci[i]] + label_cadi))
    return X_fenci,Y_fenci,label_dict
```

```python
X_fen,Y_fen = fenciClean(X,Y)
X_choose,Y_choose = choose(X_fen,Y_fen,5,20000)
```

```python
X_rule,Y_rule,dic = Rule(X_choose,Y_choose)
X_choose_again,Y_choose_again = choose(X_rule,Y_rule,5,20000)
```

```python
X_,Y_ = LR_Tfidf(X_choose_again,Y_choose_again)
```
